```python
mtches["date"]=pd.to_datetime(mtches["date"])
```

```python
mtches.dtypes
```

```python
mtches["venue_code"]=mtches["venue"].astype("category").cat.codes
```

```python
mtches["opp_code"]=mtches["opponent"].astype("category").cat.codes
```

```python
mtches["hour"]=mtches["time"].str.replace(":.+","",regex=True).astype("int")
```

```python
mtches["day_code"]=mtches["date"].dt.dayofweek
```

```python
mtches["target"]=(mtches["result"]=="W").astype("int")
```

```python
mtches
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf=RandomForestClassifier(n_estimators=50,min_samples_split=10,random_state=1)
```

```python
train=mtches[mtches["date"]<'2022-01-01']
```

```python
test=mtches[mtches["date"]>'2022-01-01']
```

```python
predictors=["venue_code","opp_code","hour","day_code"]
```

```python
rf.fit(train[predictors],train["target"],)
```

```python
preds=rf.predict(test[predictors])
```

```python
from sklearn.metrics import accuracy_score
```

```python
acc=accuracy_score(test["target"],preds)
```

```python
acc
```

```python
combined=pd.DataFrame(dict(actual=test["target"],prediction=preds))
```

```python
pd.crosstab(index=combined["actual"],columns=combined["prediction"])
```

```python
from sklearn.metrics import precision_score
```

```python
precision_score(test["target"],preds)
```
